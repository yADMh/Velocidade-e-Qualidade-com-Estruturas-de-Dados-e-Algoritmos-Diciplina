```python
def mult(x, y, depth=0):
    print("  " * depth + f"mult({x}, {y}) chamada")

    # caso base
    if y == 0:
        print("  " * depth + "retorno 0")
        return 0

    resultado = x + mult(x, y - 1, depth + 1)

    print("  " * depth + f"retorno {resultado}")
    return resultado


print("Resultado:", mult(4, 3))

# A soma acontece no retorno da recursão, porque primeiro as chamadas descem até o caso base e depois começam a somar.
```

```python
import math

def factor(x, lowest=2):
    if x == 0:
        return [0]

    if x == 1:
        return [1]

    if x < 0:
        return [-1] + factor(-x)

    for i in range(lowest, int(math.sqrt(x)) + 1):
        if x % i == 0:
            return [i] + factor(x // i, i)

    return [x]


print(factor(60))
print(factor(13))
```

```python
def power(x, y):

    if y == 0:
        return 1

    if y == 1:
        return x

    if y < 0:
        return 1 / power(x, -y)

    return x * power(x, y - 1)


print(power(2,5))
print(power(2,-3))

# O algoritmo possui complexidade O(n) porque executa y chamadas recursivas.
```

```python
import os

def list_dir(path, level=0):

    for item in os.listdir(path):

        full_path = os.path.join(path, item)

        print("  " * level + item)

        if os.path.isdir(full_path):
            list_dir(full_path, level + 1)


list_dir(".")

# Isso percorre diretórios e subdiretórios recursivamente.
```

```python
def teams(candidates, k, current="", start=0):

    if k == 0:
        print(current.strip())
        return

    for i in range(start, len(candidates)):
        teams(candidates, k-1, current + " " + candidates[i], i+1)


players = ["Ana","Bruno","Carlos","Davi"]

teams(players,2)
```

```python
def knapsack(target, weights, index=0):

    if target == 0:
        return [[]]

    if target < 0 or index >= len(weights):
        return []

    include = knapsack(target - weights[index], weights, index + 1)

    include = [[weights[index]] + c for c in include]

    exclude = knapsack(target, weights, index + 1)

    return include + exclude


weights = [2,3,4,5]

print(knapsack(7, weights))
```

### Função analisada
power(x, y)

### Observação
A função realiza uma chamada recursiva para cada redução do expoente.

### Crescimento de chamadas
Para expoente y, temos aproximadamente y chamadas.

### Complexidade
O(n)

```python
memo = {}

def power_memo(x,y):

    if y in memo:
        return memo[y]

    if y == 0:
        return 1

    if y < 0:
        return 1 / power_memo(x,-y)

    memo[y] = x * power_memo(x,y-1)

    return memo[y]


print(power_memo(2,10))
```

```python
def quicksort(arr):

    if len(arr) <= 1:
        return arr

    pivot = arr[-1]

    left = []
    right = []

    for i in arr[:-1]:

        if i <= pivot:
            left.append(i)
        else:
            right.append(i)

    return quicksort(left) + [pivot] + quicksort(right)


print(quicksort([8,3,1,7,0,10,2]))
```

```python
comparisons = 0
copies = 0

def quicksort(arr):

    global comparisons, copies

    if len(arr) <= 1:
        return arr

    pivot = arr[-1]

    left = []
    right = []

    for i in arr[:-1]:

        comparisons += 1

        if i <= pivot:
            left.append(i)
        else:
            right.append(i)

    copies += 3

    return quicksort(left) + [pivot] + quicksort(right)


data = [5,2,9,1,7]

print(quicksort(data))
print("Comparações:", comparisons)
print("Cópias:", copies)
```

QuickSort ordena completamente o conjunto de dados antes de permitir
acesso ao elemento desejado. Sua complexidade média é O(n log n).

QuickSelect utiliza a mesma lógica de partição, mas interrompe a
recursão quando o índice desejado é encontrado.

Isso reduz o trabalho desnecessário, resultando em complexidade
média O(n).

Portanto, QuickSelect é mais eficiente quando o objetivo é encontrar
apenas um elemento específico, como a mediana.
