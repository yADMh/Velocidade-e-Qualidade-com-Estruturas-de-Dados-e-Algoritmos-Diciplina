```python
# Exercicio 1

def maior_unico_quadratico(lista):
    comparacoes = 0
    maior = None
    
    for i in range(len(lista)):
        contador = 0
        for j in range(len(lista)):
            comparacoes += 1
            if lista[i] == lista[j]:
                contador += 1
        
        if contador == 1:
            if maior is None or lista[i] > maior:
                maior = lista[i]
    
    return maior, comparacoes
```

```python
# Exercicio 1 hash table

def maior_unico_hash(lista):
    acessos = 0
    frequencia = {}
    
    # Contagem
    for num in lista:
        acessos += 1
        frequencia[num] = frequencia.get(num, 0) + 1
    
    maior = None
    
    # Verificação
    for num, count in frequencia.items():
        acessos += 1
        if count == 1:
            if maior is None or num > maior:
                maior = num
    
    return maior, acessos

#Complexidade Quadrático → O(N²) |  Hash → O(N)
```

```python
# Exercicio 2

def intersecao_arrays(arr1, arr2):
    tabela = {}
    resultado = []
    
    for item in arr1:
        tabela[item] = True
    
    for item in arr2:
        if item in tabela:
            resultado.append(item)
    
    return resultado

# Complexidade: O(N + M)
```

```python
# Exercicio 3

def primeiro_duplicado(lista):
    vistos = {}
    
    for item in lista:
        if item in vistos:
            return item
        vistos[item] = True

# Complexidade: O(N)
```

```python
# Exercicio 4

import string

def letra_ausente(texto):
    texto = texto.lower()
    presentes = [False] * 26
    
    for char in texto:
        if 'a' <= char <= 'z':
            presentes[ord(char) - ord('a')] = True
    
    for i in range(26):
        if not presentes[i]:
            return chr(i + ord('a'))
        
# Complexidade Percorre texto → O(N) | Percorre 26 letras → O(1) | Resultado final: O(N)
```

```python
# Exercicio 5

def primeiro_nao_repetido(texto):
    frequencia = {}
    
    # Contagem
    for char in texto:
        frequencia[char] = frequencia.get(char, 0) + 1
    
    # Verificação
    for char in texto:
        if frequencia[char] == 1:
            return char
        
# Complexidade: O(N)
```

```python
# Exercicio 6

def selection_sort(lista):
    comparacoes = 0
    trocas = 0
    
    n = len(lista)
    
    for i in range(n):
        min_index = i
        
        for j in range(i+1, n):
            comparacoes += 1
            if lista[j] < lista[min_index]:
                min_index = j
        
        if min_index != i:
            lista[i], lista[min_index] = lista[min_index], lista[i]
            trocas += 1
    
    return lista, comparacoes, trocas

# Complexidade Sempre: O(N²)
```

```python
# Exercicio 7

def insertion_sort(lista):
    comparacoes = 0
    deslocamentos = 0
    
    for i in range(1, len(lista)):
        chave = lista[i]
        j = i - 1
        
        while j >= 0:
            comparacoes += 1
            if lista[j] > chave:
                lista[j+1] = lista[j]
                deslocamentos += 1
                j -= 1
            else:
                break
        
        lista[j+1] = chave
    
    return lista, comparacoes, deslocamentos

# Complexidade Melhor caso: O(N) | Pior caso: O(N²)
```

# Exercicio 8

Escolha entre Pilha e Fila

Sistema atende na ordem de chegada.

Estrutura correta: Fila (Queue)
Justificativa técnica

Fila implementa FIFO (First In, First Out).

Garante processamento na ordem de chegada.

Pilha é LIFO e inverteria a ordem.

Operações

enqueue (inserir)

dequeue (remover)

is_empty

is_full

```python
# Exercicio 9

class Stack:
    def __init__(self, capacidade):
        self.capacidade = capacidade
        self.dados = []
    
    def push(self, item):
        if len(self.dados) >= self.capacidade:
            raise Exception("Overflow: Pilha cheia")
        self.dados.append(item)
    
    def pop(self):
        if not self.dados:
            raise Exception("Underflow: Pilha vazia")
        return self.dados.pop()
    
    def estado(self):
        return self.dados
```

```python
# Exercicio 10

class Queue:
    def __init__(self, capacidade):
        self.capacidade = capacidade
        self.dados = [None] * capacidade
        self.inicio = 0
        self.fim = 0
        self.tamanho = 0
    
    def enqueue(self, item):
        if self.tamanho == self.capacidade:
            raise Exception("Fila cheia")
        
        self.dados[self.fim] = item
        self.fim = (self.fim + 1) % self.capacidade
        self.tamanho += 1
    
    def dequeue(self):
        if self.tamanho == 0:
            raise Exception("Fila vazia")
        
        item = self.dados[self.inicio]
        self.inicio = (self.inicio + 1) % self.capacidade
        self.tamanho -= 1
        return item
    
    def estado(self):
        return self.dados, self.inicio, self.fim, self.tamanho
```

```python
# Exercicio 11

def reverter_string(texto):
    pilha = []
    
    for char in texto:
        pilha.append(char)
    
    invertida = ""
    
    while pilha:
        invertida += pilha.pop()
    
    return invertida
```

# Exercicio 12

Análise de Funções Hash (M = 10)

Funções:

h₁(k) = k mod 10

h₂(k) = (k × 7) mod 10

h₃(k) = k² mod 10

Análise
h₁

Distribuição uniforme se dados forem bem distribuídos.
Simples e eficiente.

h₂

Multiplicação melhora dispersão para certos padrões.
Mas como 7 e 10 são coprimos, ainda mantém boa variação.

h₃

Pode gerar muitas colisões.
Exemplo:

2² mod 10 = 4

8² mod 10 = 4

Tende a repetir padrões rapidamente.

Melhor escolha geral:

h₁ ou h₂
